```python
from torch.optim import Adam
from torch_geometric.datasets import FB15k_237
from torch_geometric.nn import TransE, ComplEx
from kg_model_utils import train
```

```python
train_data = FB15k_237('data/FB15k', split='train')[0]
val_data = FB15k_237('data/FB15k', split='val')[0]
test_data = FB15k_237('data/FB15k', split='test')[0]
```

# Задача 1

```python
model = TransE(num_nodes=train_data.num_nodes,
               num_relations=train_data.num_edge_types,
               hidden_channels=50)
```

```python
loader = model.loader(head_index=train_data.edge_index[0],
                      rel_type=train_data.edge_type,
                      tail_index=train_data.edge_index[1],
                      batch_size=1000,
                      shuffle=True)
```

```python
optimizer = Adam(model.parameters(), lr=0.01)
```

```python
train(model=model,
      data_loader=loader,
      optimizer=optimizer,
      epochs=10)
```

```python
for k in [1,3,10]:
  rank, mrr, hitsN = model.test(head_index=test_data.edge_index[0],
                               rel_type=test_data.edge_type,
                               tail_index=test_data.edge_index[1],
                               batch_size=1000, k=k)
  print(f'Evaluating with k={k}')
  print(f'hits@{k}= {hitsN}')
  print(f'rank={rank}')
  print(f'mrr={mrr}')
```

# Задача 3

```python
import matplotlib.pyplot as plt
from torch.optim import Adam
from pykeen import predict
from pykeen.models import TransE
from pykeen.pipeline import pipeline
from pykeen.datasets import FB15k237, Nations
from pykeen.training import SLCWATrainingLoop
from pykeen.evaluation import RankBasedEvaluator
from pykeen.sampling import BasicNegativeSampler
```

```python
dataset = FB15k237()
```

```python
results = {}
embeddings = [40,50,100]

for e in embeddings:
  model = TransE(embedding_dim=e, triples_factory=dataset.training)
  optimizer = Adam(model.get_grad_params())

  res = pipeline(dataset=dataset,
                 model=model,
                 training_loop=SLCWATrainingLoop,
                 negative_sampler=BasicNegativeSampler,
                 evaluator=RankBasedEvaluator)

  res_df = res.metric_results.to_df()

  results[e] = res_df
```

```python
results[40]
```

```python
for e in embeddings:
  df = results[e]

  print(f"Metrics for TransE with embeding size {e}:")

  print(df[(df["Side"] == "both") & (df["Rank_type"] == 'realistic') & ((df["Metric"].str.startswith('hits_at_')) | (df["Metric"].str.startswith('inverse_harmonic_mean_rank')))])
```

 Со TransE алгоритамот од PyKeen се добиваат полоши резултати за големина на вгнездувања 40, 50, 100 + Negative Sampling

```python
accuracies = {c: 0.0 for c in countries.values()}
results = {}

for c in countries.values():
    total_predictions = 0
    correct_predictions = 0

    for r in relations_testing.values():
        true_triples = dataset.testing.triples[(dataset.testing.triples[:, 0] == c) & (dataset.testing.triples[:, 1] == r)]
        predicted_triples = preds[c][r]['tail_label'].to_list()

        for tail in predicted_triples:
            triple = [c, r, tail]
            if triple in true_triples:
                correct_predictions += 1
            total_predictions += 1

    if total_predictions > 0:
        accuracies[c] = correct_predictions / total_predictions

    results[c] = {"total_predictions": total_predictions, "accuracy": accuracies[c]}

```

```python
for c, result in results.items():
    print(f"Country: {c}, Total Predictions: {result['total_predictions']}, Accuracy: {result['accuracy']:.4f}")

print("Final Accuracies:", accuracies)
```
